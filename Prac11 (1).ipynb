```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
X = np.array([[1], [2], [3], [4], [5], [6], [7], [8]])
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])
```

```python
model = LogisticRegression()
model.fit(X, y)
```

```python
X_test = np.linspace(0,9,100).reshape(-1, 1)
y_prob = model.predict_proba(X_test)[:, 1]
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X, y, color='black', label="Data Points (Fail=0, Pass=1)")
```

```python
plt.plot(X_test, y_prob, color='blue', label="Logistic Regression Curve")
```

```python
plt.axhline(0.5, color='red', linestyle='--', label="Threshold = 0.5")
```

```python
plt.figure(figsize = (8,5))
plt.scatter(X,y , color = 'black', label = "Data points (fail = 0 , pass = 1)")
plt.plot(X_test, y_prob , color = 'blue' , label = "LogisticRegression Curve")
plt.axhline(0.5, color = 'red' , linestyle = '--', label = "Threshhold = 0.5")
plt.title("Logistic regression :study hours vs pass/fail")
plt.xlabel("Study Hours")
plt.ylabel("Probablity of passing")
plt.legend()
plt.grid(True)
plt.show()
```
